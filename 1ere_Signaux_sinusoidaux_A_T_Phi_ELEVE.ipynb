# Représentation graphiques de signaux périodiques

   **Capacité numérique** *Représenter, à l’aide d’un langage de programmation, la somme de deux signaux sinusoïdaux périodiques synchrones en faisant varier la phase à l'origine de l'un des deux ainsi que la période et l'amplitude .*

# 1er OBJECTIF


1.   Analyse des lignes de programme et Affichage d'une fonction sinusoidale


```
import numpy as np
import matplotlib.pyplot as plt

#Création d'une variable temps ,t, prenant 1000 valeurs entre 0 et 0,1 secondes#
t = np.linspace(0,0.1,100)
#La fréquence est fixée
f=100
#La déphasage phi est fixé à Pi
phi=np.pi

#Définition et affichage de la fonction sinusoidale y1
y1=3*np.sin(2*np.pi*f*t+phi)
plt.plot(t,y1,'-',color='blue')

plt.title("............................")
plt.xlabel('temps(s)')
......................
plt.grid(True)
plt.show()
```
