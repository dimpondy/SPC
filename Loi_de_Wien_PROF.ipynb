```
import pandas
import matplotlib.pyplot as plt 
data=pandas.read_csv('Wien.csv',sep=',')
x=data.longueur
y=data.temp
plt.plot(x,y)
plt.ylim(2000,20000)
plt.xlim(200,800)
plt.title('Loi de Wien')
plt.xlabel("Longueur d'onde pour le  maximum d'intensité, λmax (en nm)",fontsize=12)
plt.ylabel('température T,(en K)',fontsize=12)
plt.grid()
plt.savefig('Courbe.png')
plt.show()
```
