```
import matplotlib.pyplot as plt
import numpy
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
V=[0,2,4,6,8,10,12,13,14,15,16,17,18]
sigma=[0.68,0.67,0.65,0.64,0.62,0.61,0.60,0.65,0.70,0.74,0.79,0.84,0.88]
plt.scatter(V,sigma)
plt.grid()
plt.xlabel('V (en mL)')
plt.ylabel('σ (mS/cm)')
fit=numpy.polyfit([0,2,4,6,8,10],[0.68,0.67,0.65,0.64,0.62,0.61],1)
mod=numpy.poly1d(fit)
plt.plot([0,2,4,6,8,10,12,14],mod([0,2,4,6,8,10,12,14]),'r--')
fit_2=numpy.polyfit([13,14,15,16,17,18],[0.65,0.70,0.74,0.79,0.84,0.88],1)
mod_2=numpy.poly1d(fit_2)
plt.plot([10,13,14,15,16,17,18],mod_2([10,13,14,15,16,17,18]),'r--')
ax = plt.axes()
ax.xaxis.set_major_locator(MultipleLocator(2))

plt.show()
plt.savefig('graphe.png')
```
