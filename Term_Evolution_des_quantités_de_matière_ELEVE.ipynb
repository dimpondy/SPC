 2ème  PARTIE : Tracer les courbes APRES l'EQUIVALENCE
 
 **C-RAJOUTER** des lignes  au Script pour **afficher les courbes APRES l'EQUIVALENCE**

```
from matplotlib.pyplot import*
import matplotlib.pyplot as plt
print("Ce script permet d'afficher l'évolution des quantités  de matières \n des espèces chimiques  lors de la réaction AH + HO <=> A- + H2O")
#Initialisation des grandeurs
Cb=0.1
Va=20
VbE=18
Ca=Cb*VbE/Va
#Génération des volumes versés en base ( min, max, pas )

#AVANT l'équivalence
Vb1=np.linspace(0,VbE,10)
# Calcul des qtés de matière de A et de B qui évoluent en fct du volume de base versé
na1=(Ca*Va-Cb*Vb1)*1E-3
nb1=Cb*Vb1*1E-3
nB1=0*Vb1

#APRES l'équivalence
Vb2=np.linspace(VbE,2*VbE,10)
# Calcul des qtés de matière de A et de B
na2=0*Vb2
nb2=(0*Vb2+Cb*VbE)*1E-3
nB2=Cb*(Vb2-VbE)*1E-3
#Tracé des courbes 
plt.plot(Vb1,na1, c="r",label="AH")
plt.plot(Vb1,nb1, c="b",label="A-")
plt.plot(Vb1,nB1, c="g",label="HO-")

plt.plot(Vb2,na2, c="r")
plt.plot(Vb2,nb2, c="b")
plt.plot(Vb2,nB2, c="g")
# Légendes des graphiques
plt.title("Evolution des quantités de matières des espèces chimiques \n lors de la réaction AH + HO <=> A- + H2O")
plt.xlabel("Volume de solution titrante versé ( en mL )")
plt.ylabel("Quantités de matières des espèces \n chimiques en solution ( en mol )")
plt.grid()
plt.legend()
plt.show
```
